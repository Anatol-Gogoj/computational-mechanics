## **1. Purpose**

To establish a standardized procedure for effective tooth brushing, incorporating key metrics to ensure optimal oral hygiene and prevent dental diseases.

---

## **2. Scope**

This document applies to all individuals responsible for their oral hygiene and is intended to guide proper tooth brushing techniques using manual or electric toothbrushes.

---

## **3. Definitions**

- **Strokes Per Minute (SPM):** Number of brushing motions executed per minute.
- **Deflection of Bristles:** The bending distance of toothbrush bristles under applied pressure.
- **Quadrant:** One of the four sections of the mouth (Upper Right, Upper Left, Lower Right, Lower Left).

---

## **4. Responsibilities**

- **Users:** Adhere to this standard procedure during tooth brushing.
- **Dental Professionals:** Provide guidance and reinforce the importance of following this procedure.

---

## **5. Equipment and Materials**

- Soft-bristled toothbrush (manual or electric)
- Fluoride toothpaste (1,000–1,500 ppm fluoride)
- Dental floss
- Antimicrobial or fluoride mouthwash

---

## **6. Procedure**

### **6.1 Preparation**

1. **Toothbrush Selection**

   - Use a **soft-bristled toothbrush** with a suitable head size for comfortable reach.
   - Ensure the toothbrush is clean and in good condition (replace every **3 months** or if bristles are frayed).

2. **Toothpaste Application**

   - Apply a **pea-sized amount** (~0.25 grams) of fluoride toothpaste to the bristles.

---

### **6.2 Toothbrush Positioning**

1. **Angle of Approach**

   - Hold the toothbrush at a **45-degree angle** to the gum line.

2. **Bristle Contact**

   - Ensure bristles make contact with both the **tooth surface** and the **gum line**.

---

### **6.3 Brushing Technique**

1. **Manual Toothbrush**

   - **Strokes Per Minute:** Aim for **150–200 strokes per minute** using gentle motions.
   - **Deflection of Bristles:** Apply pressure to deflect bristles by approximately **1–2 millimeters**.
   - **Pressure Applied:** Maintain light pressure (~**100–150 grams**).

2. **Electric Toothbrush**

   - **Oscillations/Vibrations:** Utilize the brush's built-in motions (**6,000–30,000 strokes per minute**).
   - **Guidance:** Gently guide the brush over tooth surfaces without additional pressure.

---

### **6.4 Brushing Duration and Timing**

1. **Total Brushing Time**

   - Brush for a minimum of **2 minutes (120 seconds)** per session.

2. **Quadrant Allocation**

   - Divide the mouth into **4 quadrants**:

     - Upper Right
     - Upper Left
     - Lower Right
     - Lower Left

   - Spend **30 seconds** brushing each quadrant.

3. **Time Spent per Tooth Surface**

   - With **32 teeth** and **3 surfaces** each, allocate **3–4 seconds** per surface.

---

### **6.5 Coverage of Tooth Surfaces**

1. **Outer (Buccal) Surfaces**

   - Use **gentle circular motions**.

2. **Inner (Lingual/Palatal) Surfaces**

   - Use **gentle circular motions**.

3. **Chewing (Occlusal) Surfaces**

   - Use **short back-and-forth strokes**.

---

### **6.6 Additional Oral Care**

1. **Tongue Cleaning**

   - Brush the tongue for **15–20 seconds** to remove bacteria.

2. **Flossing**

   - Floss **once daily**, spending **1–2 minutes** to clean between teeth.

3. **Mouthwash Rinse**

   - Rinse with mouthwash for **30 seconds**.

---

### **6.7 Post-Brushing Procedures**

1. **Rinsing**

   - Spit out excess toothpaste; **do not rinse with water** to retain fluoride benefits.

2. **Toothbrush Maintenance**

   - Rinse toothbrush thoroughly.
   - Store upright to **air dry**.
   - **Replace** every **3 months** or when bristles are worn.

---

### **6.8 Frequency of Brushing**

1. **Daily Brushing**

   - Brush at least **twice daily**: morning and before bedtime.

2. **After Meals**

   - Brush after meals when possible, waiting **30 minutes** after consuming acidic foods or drinks.

---

## **7. Special Considerations**

1. **Toothpaste Fluoride Concentration**

   - Use toothpaste containing **1,000–1,500 ppm fluoride**.

2. **Brushing After Acidic Intake**

   - Delay brushing for **30 minutes** after acidic foods/beverages to prevent enamel erosion.

3. **Pressure Awareness**

   - Excessive force can damage gums and enamel; use gentle pressure.

---

## **8. Quality Assurance**

- **Self-Monitoring**

  - Regularly check for signs of improper brushing (e.g., gum irritation, enamel wear).

- **Professional Check-Ups**

  - Visit a dentist **twice a year** for professional cleaning and evaluation.
